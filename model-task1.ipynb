```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the dataset
data = pd.read_csv("data/hotels_europe_price.csv")
```

```python
# Calculate correlation between price and offer
corr_price_offer, p_value_offer = pearsonr(data['price'], data['offer'])
print("Correlation between price and offer:", corr_price_offer)
print("p-value:", p_value_offer)
```

```python
# Calculate correlation between price and scarce room
corr_price_scarce_room, p_value_scarce_room = pearsonr(data['price'], data['scarce_room'])
print("Correlation between price and scarce room:", corr_price_scarce_room)
print("p-value:", p_value_scarce_room)
```

```python
# Set up the plotting environment
sns.set(style="whitegrid")

# Plot for price vs. offer
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.regplot(x='offer', y='price', data=data, scatter_kws={'alpha':0.5}, line_kws={"color":"red"})
plt.title(f"Price vs. Offer\nPearson Correlation: {corr_price_offer:.2f} (p={p_value_offer:.3f})")
plt.xlabel("Offer")
plt.ylabel("Price")

# Plot for price vs. scarce room
plt.subplot(1, 2, 2)
sns.regplot(x='scarce_room', y='price', data=data, scatter_kws={'alpha':0.5}, line_kws={"color":"blue"})
plt.title(f"Price vs. Scarce Room\nPearson Correlation: {corr_price_scarce_room:.2f} (p={p_value_scarce_room:.3f})")
plt.xlabel("Scarce Room")
plt.ylabel("Price")

plt.tight_layout()
plt.show()
```
